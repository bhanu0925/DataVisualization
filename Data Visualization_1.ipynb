## Q1 Plot Histograms using iris dataset

```python
#### Ploting using matplotlib
```

```python
# ploting using matplotlib

def plot_hist(data,bins, histtype,color,title,xlable,ylable):

    plt.hist(data, bins = bins, histtype = histtype,color = color)
    plt.title(title)
    plt.xlabel(xlable)
    plt.ylabel(ylable)
    plt.show()
    
plot_hist(iris['sepal_length'],10,'stepfilled','red','Histogram of sepal Length','sepal_length_groups','length' )
plot_hist(iris['sepal_width'],20,'step','blue','Histogram of sepal Width','sepal_width_groups','length' )
plot_hist(iris['petal_length'],30,'barstacked','green','Histogram of petal Length','petal_length_groups','length' )
plot_hist(iris['petal_width'],40,'bar','black','Histogram of petal Width','petal_width_groups','length' )


```

## Q13 Plot double Bar plot for the following data also add legends

```python
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

#company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
xpos=np.arange(len(labels))
plt.xticks(xpos,labels)
plt.bar(xpos-0.2,men_means,width=0.3,label='men_means')
plt.bar(xpos+0.2,women_means,color='blue',width=0.3,label='women_means')
plt.legend()
plt.xlabel("mean")
plt.ylabel("women and men")
```

## Q14 Plot Pie Chart for the following data also add legends

```python
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)
cols = ['c', 'm', 'r', 'b']
plt.figure(figsize = (20,10))

plt.pie(sizes , labels = labels, colors = cols,   explode = (0, 0.1, 0, 0), autopct = '%1.1f%%')
plt.legend()

plt.title('Pie Plot')
plt.show()
```
